### We can Simulate Flipping a Coin Using Random Number Generation:

#### Head = 1
#### Tail = 0

```python
coin = np.random.randint(0,2)

if coin == 1:
    print("Head")
else:
    print("Tail")
```

### Simulating Flipping a Coin A Thousand Times and Counting Heads and Tails:

#### Head = 1
#### tail = 0

```python
coin = np.random.randint(0,2,1000)

heads = sum(coin)

print(f"We got Heads {heads} times.")
print(f"We got Tails {1000 - heads} times.")
```

### Rolling a Die 6000 Times and Plotting Occurance of Each Numbers:

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dice = np.random.randint(1,7,6000)

sns.countplot(x = dice)

plt.show()
```

```python
# While Rolling a Die, Probability of Each Number Ocuuring is Always 1/6.

# Here, In the Graph Above, we can see that, after Rolling a Die 6000 times, Each Number has Occurred Nearly 1000/6000 Times.
```

### If two dice are rolled, what is the probability of getting the sum of numbers greater than 7?

### Ans: 15/36 = 0.4166

```python
# Rolling Two Dice 10 Times

die1 = np.random.randint(1,7,10)
die2 = np.random.randint(1,7,10)
```

```python
sum_dice = die1 + die2
```

```python
sum_dice
```

```python
# Rolling Two Dice 100000 Times and Counting Probability of Sum Being More Than 7

die1 = np.random.randint(1,7,100000)
die2 = np.random.randint(1,7,100000)
```

```python
sum_dice = die1 + die2
```

```python
sum(sum_dice > 7)
```

```python
# As we can see, after Rolling Two Dice 100000 Times, Occurance of Sum of Two Dice more than 7 is close to Theoritical 
# Probability of 0.4166.
```
