### Simulating A Single Coin 1000 Times and Plotting it:

## Binomial Distribution Using Scipy:

```python
from scipy.stats import binom
```

### A manufacturer has 12% defects rate in production. The buyer decides to test 20 random pieces and will accept the supplier if there are 2 or less defectives. What is the probability of getting accepted?

```python
# binom.cdf(k,n,p) - Cumulative distribution function - for less than or equal to k defects

binom.cdf(2, 20, 0.12) 
```

```python
# binom.pmf(k,n,p) - Probability mass function - for exact k defects

binom.pmf(2, 20, 0.12)
```

```python
# binom.sf(k,n,p) - for more than k defects (similar to 1-cdf)

binom.sf(2, 20, 0.12)
```

```python
1 - binom.cdf(2, 20, 0.12)
```

```python
# binom.mean(n,p) - for mean of the distribution

binom.mean(20, 0.12)
```

```python
# binom.var(n,p) - for variance of the distribution

binom.var(20, 0.12)
```

```python
# binom.std(n,p) - for standard deviation of the distribution

binom.std(20, 0.12)
```

### Plotting Probability Mass Function (pmf):

### A manufacturer has 12% defects rate in production. The buyer decides to test 20 random pieces and will accept the supplier if there are 2 or less defectives. What is the probability of getting accepted?

```python
# For Our Example, Probabilty Mass Function will show Probability of Getting exactly K defective pieces while picking 
# 20 Random Pieces.
```

```python
binom.pmf(0, 20, 0.12) # Getting Exactly 0 defective Pieces.
```

```python
binom.pmf(1, 20, 0.12) # Getting Exactly 1 defective Pieces.
```

```python
binom.pmf(2, 20, 0.12) # Getting Exactly 2 defective Pieces.
```

```python
# We want to Plot The Probabily Mass Function, i.e Plotting Probability of Getting Defective Pieces = k ; 0 <= k < 20
```

```python
# For X-Axis: Number of Defective Pieces 0-20

x_axis = np.arange(0,21)
```

```python
x_axis
```

```python
# For Y-Axis: Probabily for Getting Exactly k defective Pieces.

y_axis = binom.pmf(x_axis, 20, 0.12)
```

```python
y_axis
```

```python
sns.barplot(x = x_axis, y = y_axis)
plt.show()
```

### Plotting Cumulative Density Function (cdf):

### A manufacturer has 12% defects rate in production. The buyer decides to test 20 random pieces and will accept the supplier if there are 2 or less defectives. What is the probability of getting accepted?

```python
# For Our Example, Cummulative Density Function will show Probability of Getting K or Less defective pieces while picking 
# 20 Random Pieces.
```

```python
binom.cdf(0, 20, 0.12) # Getting 0 or less defective Pieces.
```

```python
binom.cdf(1, 20, 0.12) # Getting 1 or less defective Pieces.
```

```python
binom.cdf(2, 20, 0.12) # Getting 2 or less defective Pieces.
```

```python
# We want to Plot The Cumlative Density Function, i.e Plotting Probability of Getting Defective Pieces <= k ; 0 <= k < 20
```

```python
# For X-Axis: Number of Defective Pieces 0-20

x_axis = np.arange(0,21)
```

```python
x_axis
```

```python
# For Y-Axis: Probabily for Getting k or Less defective Pieces.

y_axis = binom.cdf(x_axis, 20, 0.12)
```

```python
y_axis
```

```python
sns.barplot(x = x_axis, y = y_axis)
plt.show()
```
