## Poission Distribution Using Numpy:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Generating a random Number Using Poisson Distribution

```python
np.random.poisson(lam = 3.6) # lam = mu = average = mean of Poisson Distribution
```

```python
np.random.poisson(lam = 3.6)
```

#### Generating 10 random Numbers Using Poisson Distribution

```python
np.random.poisson(3.6, 10)
```

### On a booking counter on the average 3.6 people come every 10 minutes on weekends. What is the probability of getting 7 people in 10 minutes?

```python
# mu = lam = mean = 3.6
```

```python
# Generating 10000 Random Numbers Using Poisson Distribution

queue = np.random.poisson(3.6, 10000)

sns.countplot(x = queue)
plt.show()
```

```python
# We are interested in Probability of Getting 7 People in Queue.

sum(queue == 7)
```

```python
sum(queue == 7) / 10000
```

```python
# Answer Matches Theoritical Probability of Getting 7 People in Queue.
```

## Poission Distribution Using Scipy:

```python
from scipy.stats import poisson
```

```python
# poisson.cdf(k,mu) - Cumulative distribution function - for less than or equal to 

poisson.cdf(7, 3.6)
```

```python
# poisson.pmf(k,mu) - Probability mass function - for exact value

poisson.pmf(7, 3.6)
```

```python
# poisson.sf(k,mu) - for more than (similar to 1-cdf)

poisson.sf(7, 3.6)
```

```python
1 - poisson.cdf(7, 3.6)
```

```python
# poisson.mean(mu) - for mean of the distribution

poisson.mean(3.6)
```

```python
# poisson.var(mu) - for variance of the distribution

poisson.var(3.6)
```

```python
# poisson.std(mu) - for standard deviation of the distribution

poisson.std(3.6)
```

### Plotting Probability Mass Function (pmf) and Cumulative Density Function (cdf):

```python
# For x-Axis

x_axis = np.arange(0,15)
```

```python
x_axis
```

```python
# For y_Axis

y_axis = poisson.pmf(x_axis, 3.6)
```

```python
y_axis
```

```python
sns.barplot(x = x_axis, y = y_axis)
plt.show()
```

```python
y_axis = poisson.cdf(x_axis, 3.6)
```

```python
y_axis
```

```python
sns.barplot(x = x_axis, y = y_axis)
plt.show()
```
