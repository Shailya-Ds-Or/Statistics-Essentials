## Normal Distribution Using Numpy:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Generating a Random Number Using Normal Distribution with Mean = 200, Standard Deviation = 2

```python
np.random.normal(100, 2)
```

```python
np.random.normal(100, 2)
```

#### Generating 10 Random Numbers Using Normal Distribution with Mean = 200, Standard Deviation = 2

```python
np.random.normal(100, 2, 10)
```

#### Generating 1000 Random Numbers Using Normal Distribution with Mean = 200, Standard Deviation = 2 an Plotting:

```python
normal_dist = np.random.normal(100, 2, 1000)

sns.displot(data= normal_dist)
plt.show()
```

#### Generating 100000 Random Numbers Using Normal Distribution with Mean = 200, Standard Deviation = 2 an Plotting:

```python
normal_dist = np.random.normal(100, 2, 100000)

sns.displot(data= normal_dist)
plt.show()
```

```python
normal_dist = np.random.normal(100, 2, 100000)

sns.displot(data= normal_dist, kde= True)
plt.show()
```

```python
# Checking How many values are Within 1 Standard Deviation of the Mean

sum((normal_dist > 98) & (normal_dist < 102))
```

```python
# Checking How many values are Within 2 Standard Deviation of the Mean

sum((normal_dist > 96) & (normal_dist < 104))
```

```python
# Checking How many values are Within 3 Standard Deviation of the Mean

sum((normal_dist > 94) & (normal_dist < 106))
```

## Normal Distribution Using Scipy:

```python
from scipy.stats import norm
```

```python
# norm.cdf(x,mu,sigma) - Cumulative distribution function - for less than or equal to 

norm.cdf(102, 100, 2)
```

```python
# norm.pdf(x,mu,sigma) - Probability density function (not Probability mass function) - for exact value

norm.pdf(102, 100, 2)
```

```python
# norm.sf(x,mu,sigma) - for more than (similar to 1-cdf)

norm.sf(102, 100, 2)
```

```python
1 - norm.cdf(102, 100, 2)
```

```python
# norm.mean(mu) - for mean of the distribution

norm.mean(100)
```

```python
# norm.var(mu) - for variance of the distribution

norm.var(100)
```

```python
# norm.std(mu) - for standard deviation of the distribution

norm.std(100)
```

### Question: Perfume bottles are filled with the average volume of 150 cc and the standard deviation of 2 cc. What percent of bottles will have a volume greater than 153 cc?

```python
# mu= 150
# std = 2
# We want to find norm.sf(153, 150, 2) or 1 - norm.cdf(153, 150, 2)
```

```python
norm.sf(153, 150, 2)
```

```python
1 - norm.cdf(153, 150, 2)
```

### Question: Perfume bottles are filled with the average volume of 150 cc and the standard deviation of 2 cc. What percent of bottles will have a volume between 148 and 152 cc?

```python
# mu = 150
# std = 2

# Here, we want to find Probability Between values of 148 and 152.
# So, we Can Find norm.sf(148, 150, 2) and subtract norm.sf(152, 150, 2) from it.
# OR
# We can Find norm.cdf(152, 150, 2) and subtract norm.cdf(148, 150, 2) from it.
```

```python
norm.sf(148, 150, 2) - norm.sf(152, 150, 2)
```

```python
norm.cdf(152, 150, 2) - norm.cdf(148, 150, 2)
```
