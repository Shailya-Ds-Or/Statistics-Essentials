## Central Limit Theorem Using Python:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Central Limit Theorem Using One Population:

#### Rolling a Die 10,00,000 times and Plotting:

```python
die_million = np.random.randint(1,7,1000000)
```

```python
die_million
```

```python
sns.displot(die_million, bins= 6)
plt.show()
```

```python
# As we can see, this is a Uniform Distribution.
```

```python
np.mean(die_million) # Mean Of Population
```

```python
np.std(die_million) # Standard Deviation of Population
```

#### Drawing Samples of Size 4 from Population:

```python
die_million.reshape(250000,4)
```

#### Storing Means of These Samples and Plotting:

```python
mean_4_samples = die_million.reshape(250000,4).mean(axis= 1) # Axis=1 meaning we are taking mean across columns
```

```python
mean_4_samples
```

```python
sns.displot(mean_4_samples, bins= 6)
plt.show()
```

```python
# As we can see, Distribution of Sample Means Closely Resembles Normal Distribution.
```

```python
np.mean(mean_4_samples) # Mean of Smaple Mean is Close to Population Mean.
```

```python
np.std(mean_4_samples) # Standard Deviation of Sample Means is Close to Population Mean / Square Root of Sample size
```

#### Drawing Samples of Size 100 from Population:

```python
die_million.reshape(10000,100)
```

#### Storing Means of These Samples and Plotting:

```python
mean_100_samples = die_million.reshape(10000,100).mean(axis= 1) # Axis=1 meaning we are taking mean across columns
```

```python
np.std(mean_100_samples) # Standard Deviation of Sample Means is Close to Population Mean / Square Root of Sample Size
```

### Central Limit Theorem Using Two Populations:

```python
var1 = np.random.normal(100, 5, 10000) # np.random.norman(mean, std, size)
```

```python
var2 = np.random.normal(80, 5, 10000) # np.random.norman(mean, std, size)
```

```python
var1
```

```python
var2
```

#### Mixing these two Populations and Shuffling the Mixed Data:

```python
comb_data = np.append(var1, var2)
```

```python
comb_data
```

```python
np.random.shuffle(comb_data)
```

```python
comb_data
```

```python
sns.displot(comb_data)
plt.show()
```

```python
np.mean(comb_data)
```

```python
np.std(comb_data)
```

#### Drawing Samples of Size 2 from Combined Data:

```python
comb_data.reshape(10000,2)
```

#### Storing Means of These Samples and Plotting:

```python
mean_2_comb_data = comb_data.reshape(10000,2).mean(axis= 1)
```

```python
mean_2_comb_data
```

```python
sns.displot(mean_2_comb_data)
plt.show()
```

```python
np.mean(mean_2_comb_data)
```

```python
np.std(mean_2_comb_data)
```
