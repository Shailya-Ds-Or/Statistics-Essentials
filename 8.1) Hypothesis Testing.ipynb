## Hypothesis Testing Using Python:

```python
import numpy as np
from scipy.stats import norm
```

A machine is producing perfume bottles with the long term average of 150cc and the standard deviation of 2 cc.
Four bottles were picked and the average volume was found to be 153.8cc.

Has the average volume changed from 150cc? Test with 95% confidence level.

Alternate Hypothesis: Mean volume != 150cc

Null Hypothesis: Mean volume = 150cc

Alpha = 1 - 0.95 = 0.05

Z calculated = (Observed mean - population mean) / standard error

z = (x — μ) / (σ / √n)

```python
z_calculated = (153.8 - 150) / (2 / np.sqrt(4))
```

```python
z_calculated
```

##### norm.pdf(x, loc=0, scale=1) Probability density function.

#### norm.cdf(x, loc=0, scale=1)  Cumulative distribution function. (area to the left of x)

#### norm.sf(x, loc=0, scale=1) Survival function (also defined as 1 - cdf, but sf is sometimes more accurate).

#### norm.ppf(q, loc=0, scale=1) Percent point function (inverse of cdf — percentiles).

####  norm.isf(q, loc=0, scale=1) Inverse survival function (inverse of sf).

```python
# We have alpha = 0.025
# We have Two-Tailed Test

# So, we need to find Critical Values on both sides of Tails with area of 2.5% on each side.
```

```python
# norm.cdf gives area to the left of given point in distribution.
# norm.ppf is inverse of norm.cdf, giving point in Distribution based on area to the left.

norm.ppf(0.025)
```

```python
# norm.sf gives area to the right of given point in distribution.
# norm.isf is inverse of norm.sf, giving point in Distribution based on area to the right.
```

```python
norm.isf(0.025)
```

```python
# So, Critical Value of 1.96 < calculated value of Z (3.8).

# Hence, We Can Reject Null Hypothesis.
```

```python
p_value = 2 * norm.sf(z_calculated)
```

```python
p_value
```

```python
# p-Value < alpha.

# Hence, We can Reject Null Hypothesis.
```

```python
p_value = (1 - norm.cdf(z_calculated)) * 2
```

```python
p_value
```

```python
# p-Value < alpha.

# Hence, We can Reject Null Hypothesis.
```
