### Goodness of Fit Test - Chi Square Test Using Python:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

#### Example1:

A coin is flipped 100 times. Number of heads and tails are noted. Is this coin biased?
Check with 95% Confidence Level.

Heads = 40

Tails = 60

Ho: Coin is not biased.

Ha: Coin is biased.

Alpha = 0.05

```python
observed_data = [40 , 60]
```

```python
expected_data = [50 , 50]
```

```python
chi_calculated, p_value = stats.chisquare(f_obs= observed_data, f_exp= expected_data)
```

```python
print(chi_calculated)
print(p_value)
```

```python
# p-Value = 0.04

# p-Value < alpha

# Hence, We Can Reject The Null Hypothesis.
```

#### Example2:

A die is rolled 65 times. Number of occurance of each number are noted. Is this die biased?
Check with 95% Confidence Level.

1 = 10

2 = 6

3 = 8

4 = 22

5 = 11

6 = 8

Ho: Die is not biased.

Ha: Die is biased.

Alpha = 0.05

```python
die_observed = [10, 6, 8, 22, 11, 8]
```

```python
chi_calculated, p_value = stats.chisquare(f_obs= die_observed)
```

```python
print(chi_calculated)
print(p_value)
```

```python
# p-Value = 0.009

# p-Value < alpha

# Hence, We Can Reject The Null Hypothesis.
```
