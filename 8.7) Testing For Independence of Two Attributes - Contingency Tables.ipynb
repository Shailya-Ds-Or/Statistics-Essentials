### Contingency Tables Using Python:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

#### Example1:

H0 : There is no relationship between the row and column variables.

Ha : There is a relationship.

alpha = 0.05

```python
shift_and_operator_df = pd.DataFrame(data = np.array([[22,26,23],[28,62,26],[72,22,66]]),
                                    columns= ["Operator 1", "Operator 2", "Operator 3"],
                                    index= ["Shift 1", "Shift 2", "Shift 3"])
```

```python
shift_and_operator_df
```

```python
stats.chi2_contingency(observed= shift_and_operator_df)
```

```python
# p-Value = 0.0000...

# p-Value < alpha

# Hence, We can Reject The Null Hypothesis.
```

#### Example2:

In Tips Dataset from Seaborn Library, Check if There is a relationship between Smoking Habit and Day of the Week.

H0 : There is No Relationship Between Smoking Habits and Days of the Weeks.

Ha: There is a Relationship Between Smoking Habit and Days of Weeks.

alpha = 0.05

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
tips.shape
```

```python
tips.info()
```

```python
tips.describe()
```

```python
smk_and_days = pd.pivot_table(data = tips, index= "day", columns= "smoker", aggfunc= "count")["tip"]
```

```python
smk_and_days
```

```python
stats.chi2_contingency(observed= smk_and_days)
```

```python
# p_Value = 0.000...

# p_Value < alpha

# Hence, We Can Reject The Null Hypothesis.
```
